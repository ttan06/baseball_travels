```python
def home_game_exists(schedule, # pandas data frame
                     team, # string
                     start_date, #datetime
                     end_date #datetime
                    ):
    schedule.loc[(schedule['date'] >= start_date) & (schedule['date'] <= end_date)]
    for home_team in schedule['home team']:
        if home_team == team:
            return True
    return False
```
